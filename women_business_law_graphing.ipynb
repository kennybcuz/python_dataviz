### Import Statements

```python
# import statements 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

### Load Data

```python
# import data into dataframe
filename = r"D:\CodingDojo\Python_DataViz\WBL50YearPanelDetailsWeb01Jun2020.xls"

# create dataframe
df = pd.read_excel(filename, sheet_name='WBL1971-2020', header= 0,)
```

### Change column names

```python
# inspect column names
df.columns
```

```python
# a dictionary of column categories with questions
# in case I want to swap column categories with column questions
new_cols = df.iloc[0,:]
dictCols = dict(zip(new_cols.index,new_cols.values))
dictCols
```

```python
# create a dictionary for Unnamed columns

dictUnnamed = dict()
# Iterate over all the items in dictionary and filter items which has even keys
for (key, value) in dictCols.items():
   # Check if key is even then add pair to new dictionary
   if 'Unnamed:' in key:
       dictUnnamed[key] = value

print(dictUnnamed)
```

```python
# rename Unnamed columns
df.rename(dictUnnamed, axis=1, inplace=True)
```

```python
# verify column name change
df.columns
```

```python
# drop one of the header rows
# split into dataframes with different columns selected

df1 = df.drop(0, axis=0).reset_index(drop=True)
df2 = df.rename(columns=df.iloc[0,:]).drop(0, axis=0).reset_index(drop=True)
df3 = df[[]]
```

### Change Column data types

```python
# change column data types
# use df2

numeric = ['WBL Report Year',
           'WBL INDEX','MOBILITY',
           'WORKPLACE',
           'PAY',
           'MARRIAGE',
           'PARENTHOOD',
           'ENTREPRENEURSHIP',
           'ASSETS',
           'PENSION']

df2[numeric] = df2[numeric].apply(pd.to_numeric)
```

# Visualizations

## Inequality towards Women by Income Group

```python
# set the figure size
sb.set_theme(rc = {'figure.figsize': (20,20), 'axes.labelsize': 14})
sb.set_style('whitegrid')

# create plots
g = sb.relplot(x="WBL Report Year", 
               y="WBL INDEX", 
               hue="Region",
               units="Economy", 
               style=None, 
               size=None, 
               estimator=None,
               kind="line", 
               col="Income group",
               col_wrap=2,
               row=None,
               height=4,
               aspect=2,
               linewidth=0.5,
               data=df2);

# set figure title
g.fig.suptitle('Inequality towards Women by Income Group',
              fontsize= 18,
              fontweight= 'bold')

# adjust overlap between axes titles and figure titles
g.fig.subplots_adjust(top= 0.90)

# set axis labels for axes
g.set_axis_labels('Year','WBL Equality Index')

# show plot
plt.show()
```

```python
# set the figure size
sb.set_theme(rc = {'figure.figsize': (20,10), 'axes.labelsize': 14})
sb.set_style('whitegrid')

# create plots
g = sb.catplot(x="Income group", 
               y="WBL INDEX", 
               hue=None,
               kind="bar", 
               col=None,
               col_wrap=None,
               row=None,
               height=4,
               aspect=2,
               data=df2);

# set figure title
g.fig.suptitle('Inequality towards Women by Income Group',
              fontsize= 18,
              fontweight= 'bold')

# adjust overlap between axes titles and figure titles
g.fig.subplots_adjust(top= 0.90)

# set axis labels for axes
g.set_axis_labels('Income Group','WBL Equality Index')

# show plot
plt.show()
```

```python
# set the figure size
sb.set(rc = {'figure.figsize': (20,20),
            'axes.labelsize': 14})

# set plot style
sb.set_style('whitegrid')

# create plots
g = sb.catplot(x="Income group",
            y="WBL INDEX",
            hue=None,
            kind="violin",
            inner="stick",
            split=False,
            palette="pastel",
            height=10,
            aspect=2,
            data=df2)

# set figure title
g.fig.suptitle('Inequality towards Women by Income Group',
              fontsize= 18,
              fontweight= 'bold')

# adjust overlap between axes titles and figure titles
g.fig.subplots_adjust(top= 0.90)

# set axis labels for axes
g.set_axis_labels('Year','WBL Equality Index')

# show plot
plt.show()
```

```python
# set the figure size
sb.set_theme(rc = {'figure.figsize': (20,10), 'axes.labelsize': 14})
sb.set_style('whitegrid')

# create subplot figure and title
fig, axes = plt.subplots(1,2)
fig.suptitle('Inequality Towards Women Still a Challenge for some Upper & High Income Countries', 
             fontsize= 18, 
             fontweight= 'bold')

# create bar plot
sb.barplot(ax= axes[0],
               x="Income group", 
               y="WBL INDEX", 
               hue=None,
               data=df2);



# create violin plot
sb.violinplot(ax= axes[1],
            x="Income group",
            y="WBL INDEX",
            hue=None,
            kind="violin",
            inner="stick",
            split=False,
            palette="pastel",
            data=df2)


# show plot
plt.show()
```

## Inequality towards Women by Life Events

```python
# set the figure size
sb.set_theme(rc = {'figure.figsize': (20,20), 'axes.labelsize': 14})
sb.set_style('whitegrid')

fig, axes = plt.subplots(3, 3)
fig.suptitle('Inequality Towards Women by Life Events', 
             fontsize= 18, 
             fontweight= 'bold')

# create plots
sb.violinplot(ax= axes[0,0],
            x="Income group",
            y="MOBILITY",
            hue=None,
            kind="violin",
            inner="stick",
            split=False,
            palette="pastel",
            data=df2)

sb.violinplot(ax= axes[0,1],
            x="Income group",
            y="WORKPLACE",
            hue=None,
            kind="violin",
            inner="stick",
            split=False,
            palette="pastel",
            data=df2)

sb.violinplot(ax= axes[0,2],
            x="Income group",
            y="PAY",
            hue=None,
            kind="violin",
            inner="stick",
            split=False,
            palette="pastel",
            data=df2)

sb.violinplot(ax= axes[1,0],
            x="Income group",
            y="MARRIAGE",
            hue=None,
            kind="violin",
            inner="stick",
            split=False,
            palette="pastel",
            data=df2)

sb.violinplot(ax= axes[1,1],
            x="Income group",
            y="PARENTHOOD",
            hue=None,
            kind="violin",
            inner="stick",
            split=False,
            palette="pastel",
            data=df2)

sb.violinplot(ax= axes[1,2],
            x="Income group",
            y="ENTREPRENEURSHIP",
            hue=None,
            kind="violin",
            inner="stick",
            split=False,
            palette="pastel",
            data=df2)

sb.violinplot(ax= axes[2,0],
            x="Income group",
            y="ASSETS",
            hue=None,
            kind="violin",
            inner="stick",
            split=False,
            palette="pastel",
            data=df2)

sb.violinplot(ax= axes[2,1],
            x="Income group",
            y="PENSION",
            hue=None,
            kind="violin",
            inner="stick",
            split=False,
            palette="pastel",
            data=df2)

# adjust plot overlaps
plt.subplots_adjust(bottom=0.15, wspace=0.20)

# show plot
plt.show()
```
